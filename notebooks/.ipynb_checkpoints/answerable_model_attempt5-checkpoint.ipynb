```python
import utils
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline, FeatureUnion
from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression, LinearRegression, Perceptron
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import cross_val_score, KFold, RandomizedSearchCV, GridSearchCV, cross_val_predict
from sklearn.base import BaseEstimator, TransformerMixin
from nltk import word_tokenize
from nltk.corpus import stopwords
import pandas as pd
from sklearn.metrics import precision_score, accuracy_score, recall_score, f1_score, classification_report, make_scorer, roc_auc_score, confusion_matrix, SCORERS
from model_training import WordCountTransformer, CharacterCountTransformer, StopWordFrequency
from sklearn.neighbors import KNeighborsClassifier

stop_words = set(stopwords.words('English'))
```

```python
def answerable_preprocessor(sentence):
    sentence = utils.check_quotations(sentence)
    sentence = word_tokenize(sentence)
    
    for i in range(len(sentence)):
        sentence[i] = utils.check_camel_snakecase(sentence[i])
        sentence[i] = sentence[i].lower()
        
#         if sentence[i] in stop_words and sentence[i] not in ['who', 'what', 'when', 'where', 'why', 'how']:
#             sentence[i] = ""
        if sentence[i].isnumeric():
            sentence[i] = "numericnumber"
        elif utils.is_spelled_out_number(sentence[i]):
            sentence[i] = "nonnumericnumber"
        elif utils.is_TA_or_instructor_name(sentence[i]):
            sentence[i] = "name"
        elif utils.is_error_code(sentence[i]):
            sentence[i] = "errorcode"
        elif utils.is_system_word(sentence[i]):
            sentence[i] = "sys"
        elif utils.is_function(sentence[i]):
            sentence[i] = "func"

    return ' '.join(sentence)

def partition_data(data, label):
    cpy = data.copy()
    
    for i in range(len(data)):
        if data[i][1] == label:
            cpy[i][1] = 1
        else:
            cpy[i][1] = 0
    
    return cpy
```

<h3>T - Is it a clear question/Is there enough information for TA to understand what is happening?</h3> <br />
<h3>F - It is not a clear question, does not provide enough information, or simply asks for answer</h3> <br />
<h3>Frustrated - Anything that indicates the student is confused and cannot provide any more detail. "help explaining general thing"</h3>

```python
answerable_data = utils.load_data('../data/tot.json', 'answerable')
t_f_data = np.array([[answerable_data[:,0][i], answerable_data[:,1][i]] for i in range(len(answerable_data)) if answerable_data[:,1][i] == 't' or answerable_data[:,1][i] == 'f'])
c_f_data = np.array([[answerable_data[:,0][i], answerable_data[:,1][i]] for i in range(len(answerable_data)) if answerable_data[:,1][i] == 'c' or answerable_data[:,1][i] == 'f'])
t_c_data = np.array([[answerable_data[:,0][i], answerable_data[:,1][i]] for i in range(len(answerable_data)) if answerable_data[:,1][i] == 't' or answerable_data[:,1][i] == 'c'])
t_data = partition_data(answerable_data, 't')
c_data = partition_data(answerable_data, 'c')
f_data = partition_data(answerable_data, 'f')
d = np.array([[answerable_data[:,0][i], 't'] if (answerable_data[:,1][i] == 't' or answerable_data[:,1][i] == 'c') else [answerable_data[:,0][i], 'f'] for i in range(len(answerable_data))])
```

# Model 1: Multiclass Logistic Regression C-Val Acc: 85%

```python
pipeline = Pipeline([
    ('features', FeatureUnion([
        ('vectorizer', TfidfVectorizer(preprocessor=answerable_preprocessor)),
        ('char_count', CharacterCountTransformer()),
        ('word_count', WordCountTransformer())
    ])),
#     ('clf', LogisticRegression(max_iter=1000, class_weight={'t':2.5}))
    ('clf', LogisticRegression(max_iter=1000, class_weight={'t': 2.5}))
])

pipeline.fit(d[:,0], d[:,1])
np.mean(cross_val_score(pipeline, d[:,0], d[:,1]))
```

```python
y_pred = cross_val_predict(pipeline, d[:,0], d[:,1], cv=5)
conf_mat = confusion_matrix(d[:,1], y_pred, labels=['t', 'f'])

df_cm = pd.DataFrame(conf_mat, index = [i for i in ['t', 'f']],
                  columns = [i for i in ['t', 'f']])
plt.figure(figsize = (10,7))
sn.heatmap(df_cm, annot=True)
```

```python
param_grid = {
    'features__vectorizer__ngram_range': [(1,1), (1,2), (1,3)],
    'features__vectorizer__max_features': [None, 100, 250, 500, 1000, 1500, 2000, 2500, 3000],
    'clf__C': [.01, .1, 1, 10, 100],
    'clf__class_weight': [{'t': 2}, {'t': 2.5}, {'t': 1.5}, {'t': 3}, {'f': .5}, {'f': .75}]
}

search = RandomizedSearchCV(pipeline, param_grid, n_jobs=-1, n_iter=100)
search.fit(d[:,0], d[:,1])
np.mean(cross_val_score(search.best_estimator_, d[:,0], d[:,1], cv=5))
```

### Putting Together

```python
from collections import Counter

kf = KFold(n_splits=5)
cvs = []

for train,test in kf.split(answerable_data):
    train = answerable_data[train]
    test = d[test]
    
    t_f_data = np.array([[train[:,0][i], train[:,1][i]] for i in range(len(train)) if train[:,1][i] == 't' or train[:,1][i] == 'f'])
    c_f_data = np.array([[train[:,0][i], train[:,1][i]] for i in range(len(train)) if train[:,1][i] == 'c' or train[:,1][i] == 'f'])
    
    t_f_search.best_estimator_.fit(t_f_data[:,0], t_f_data[:,1])
    c_f_search.best_estimator_.fit(c_f_data[:,0], c_f_data[:,1])
    
    t_f_preds = t_f_search.best_estimator_.predict(test[:,0])
    c_f_preds = c_f_search.best_estimator_.predict(test[:,0])
    
    tot_preds = []
    
    for i in range(len(t_f_preds)):
        preds = [t_f_preds[i], c_f_preds[i]]
        
        if preds == ['t', 'c'] or preds == ['t', 'f']:
            tot_preds.append('t')
        elif preds == ['f', 'c']:
            tot_preds.append('t')
        elif preds == ['f', 'f']:
            tot_preds.append('f')
            
    cvs.append(accuracy_score(test[:,1], tot_preds))
    conf_mat = confusion_matrix(test[:,1], tot_preds, labels=['t', 'f'])

    df_cm = pd.DataFrame(conf_mat, index = [i for i in ['t', 'f']],
                      columns = [i for i in ['t', 'f']])
    plt.figure(figsize = (7,4))
    sn.heatmap(df_cm, annot=True)
    plt.show()
    print(classification_report(test[:,1], tot_preds))
    
```

```python
from collections import Counter

kf = KFold(n_splits=5)
cvs = []

for train,test in kf.split(answerable_data):
    train = answerable_data[train]
    test = answerable_data[test]
    
    t_data = partition_data(train, 't')
    c_data = partition_data(train, 'c')
    f_data = partition_data(train, 'f')
    
    t_search.best_estimator_.fit(t_data[:,0], t_data[:,1])
    c_search.best_estimator_.fit(c_data[:,0], c_data[:,1])
    f_search.best_estimator_.fit(f_data[:,0], f_data[:,1])
    
    t_preds = t_search.best_estimator_.predict_proba(test[:,0])[:,1]
    c_preds = c_search.best_estimator_.predict_proba(test[:,0])[:,1]
    f_preds = f_search.best_estimator_.predict_proba(test[:,0])[:,1]
    
    tot_preds = []
    
    for i in range(len(t_preds)):
        preds = [t_preds[i], c_preds[i], f_preds[i]]
        if max(preds) == c_preds[i]:
            tot_preds.append('c')
        elif max(preds) == t_preds[i]:
            tot_preds.append('t')
        elif max(preds) == f_preds[i]:
            tot_preds.append('f')
        
    cvs.append(accuracy_score(test[:,1], tot_preds))
    conf_mat = confusion_matrix(test[:,1], tot_preds, labels=['t', 'f', 'c'])

    df_cm = pd.DataFrame(conf_mat, index = [i for i in ['t', 'f', 'c']],
                      columns = [i for i in ['t', 'f', 'c']])
    plt.figure(figsize = (7,4))
    sn.heatmap(df_cm, annot=True)
    plt.show()
    print(classification_report(test[:,1], tot_preds))
    
```

```python
np.mean(cvs)
```

# Using a Classifier to Blend Preds C-Val: 75.5%

```python
from collections import Counter

kf = KFold(n_splits=5)
cvs = []

for train,test in kf.split(answerable_data):
    train = answerable_data[train]
    test = answerable_data[test]
    
    t_f_data = np.array([[train[:,0][i], train[:,1][i]] for i in range(len(train)) if train[:,1][i] == 't' or train[:,1][i] == 'f'])
    c_f_data = np.array([[train[:,0][i], train[:,1][i]] for i in range(len(train)) if train[:,1][i] == 'c' or train[:,1][i] == 'f'])
    t_c_data = np.array([[train[:,0][i], train[:,1][i]] for i in range(len(train)) if train[:,1][i] == 't' or train[:,1][i] == 'c'])
    
    t_f_search.best_estimator_.fit(t_f_data[:,0], t_f_data[:,1])
    c_f_search.best_estimator_.fit(c_f_data[:,0], c_f_data[:,1])
    t_c_search.best_estimator_.fit(t_c_data[:,0], t_c_data[:,1])
    
    t_f_train_pred = t_f_search.best_estimator_.predict(train[:,0])
    c_f_train_pred = c_f_search.best_estimator_.predict(train[:,0])
    t_c_train_pred = t_c_search.best_estimator_.predict(train[:,0])
    key = {
        't': 1,
        'f': 0,
        'c': .75
    }
    x = [[key[t_f_train_pred[i]], key[c_f_train_pred[i]], key[t_c_train_pred[i]]] for i in range(len(train))]
    
    clf = LogisticRegression()
    clf.fit(x, train[:,1])
    
    t_f_test_pred = t_f_search.best_estimator_.predict(test[:,0])
    c_f_test_pred = c_f_search.best_estimator_.predict(test[:,0])
    t_c_test_pred = t_c_search.best_estimator_.predict(test[:,0])
    
    x_test = [[key[t_f_test_pred[i]], key[c_f_test_pred[i]], key[t_c_test_pred[i]]] for i in range(len(test))]
    
    tot_preds = clf.predict(x_test)
            
    cvs.append(accuracy_score(test[:,1], tot_preds))
    conf_mat = confusion_matrix(test[:,1], tot_preds, labels=['t', 'f', 'c'])

    df_cm = pd.DataFrame(conf_mat, index = [i for i in ['t', 'f', 'c']],
                      columns = [i for i in ['t', 'f', 'c']])
    plt.figure(figsize = (7,4))
    sn.heatmap(df_cm, annot=True)
    plt.show()
    print(classification_report(test[:,1], tot_preds))
    
```

```python
np.mean(cvs)
```
